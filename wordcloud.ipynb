```python
import random
from konlpy.tag import *
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline

f = open("gas.txt") # gas 파일 읽어옴
text = f.read()

okt = Okt() # konlpy okt 불러옴
nouns = okt.nouns(text) # 명사단위로 
nouns = [n for n in nouns if len(n) > 1] # 한글자 단어는 제외

count = Counter(nouns) # 개수를 세줌
font_path = "c:/Windows/Fonts/malgun.ttf" # 한글 폰트 경로

tags = count.most_common(200) # 빈도수 상위 200개만 출력
wc = WordCloud(font_path=font_path, background_color='white', width=1000, height=800) # 너비 밑 기타 속성
cloud = wc.generate_from_frequencies(dict(tags))
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
f.close()
```
